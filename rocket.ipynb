```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("amazon_cleaned.csv")
```

```python
df.head()
df.shape
df.columns
df.info()
```

TO check missing values

```python
df.isnull().sum()
```

Check duplicate rows

```python
df.duplicated().sum() #it returns the total count of duplicate rows as duplicated() just return T or F and then sum gives the total sum of T or F
```

Remove duplicate

```python
df.drop_duplicates(inplace=True)
```

```python
df['rating'].unique()
```

Clean Rating Column
Convert invalid values into numeric:

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

Handle Missing Rating Values

```python
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

market price estimate

```python
df['market_price'] = (df['actual_price'] + df['discounted_price']) / 2
```

Seller margin

```python
df['seller_margin'] = df['actual_price'] - df['discounted_price']
```

Estimated Demand Score

```python
df['demand_score'] = df['rating'] * df['rating_count']
```

Simulated Competitor Price

```python
df['competitor_price'] = df['discounted_price'] * 1.05
```

Product Length Feature

```python
df['description_length'] = df['about_product'].apply(len)
```

Brand Extraction

```python
df['brand'] = df['product_name'].apply(lambda x: x.split()[0])
```

```python
df.to_csv("amazon_rocketsales.csv", index=False)
```
